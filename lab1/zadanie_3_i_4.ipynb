```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: x**2 + 5
first_range = np.linspace(-1, 1, 100)
second_range = np.linspace(-6, 6, 100)
third_range = np.linspace(-0, 5, 100)

plt.plot(first_range, f(first_range), "r-", label="pierwszy zakres")
plt.xlabel('x')
plt.ylabel('y')
plt.title('wykres')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(second_range, f(second_range), "g-", label="drugi zakres")
plt.xlabel('x')
plt.ylabel('y')
plt.title('wykres')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(third_range, f(third_range), "y-", label="trzeci zakres")
plt.xlabel('x')
plt.ylabel('y')
plt.title('wykres')
plt.grid()
plt.legend()
plt.show()
```
